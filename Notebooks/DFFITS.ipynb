```python
from sklearn.linear_model import LinearRegression
import pandas as pd
from tabulate import tabulate
from rich import print as rprint

# Path to the uploaded CSV file
csv_path = './data.csv'
# Read the data from the CSV file
df = pd.read_csv(csv_path)

# Define predictor (X) and response (y)
# Assuming the column names in the CSV are 'X' for the predictor and 'y' for the response
X = df[['X']].values.reshape(-1, 1)
y = df['y'].values

# Create and fit the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions using the model
y_pred = model.predict(X)

# Prepare a DataFrame with the original data and predictions for display
results_df = pd.DataFrame({
    'Prädikator_X': df['X'],
    'Kriterium_y': df['y'],
    'y_pred': y_pred
})

# Print the table using tabulate
formatted_table = tabulate(results_df, headers='keys', tablefmt='presto', numalign="right", showindex=False)
rprint(formatted_table)
```
